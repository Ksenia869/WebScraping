```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
import json
```

```python
mail = 'ksenia98870@yandex.ru'
with open('password.txt','r') as fil:
    password = fil.readline()
```

```python
driver = webdriver.Chrome()

driver.get("https://mail.yandex.ru/")
WebDriverWait(driver,120)

login = driver.find_element(By.ID, 'header-login-button')
login.click()
WebDriverWait(driver, 120).until(expected_conditions.presence_of_all_elements_located((By.ID, 'passp-field-login')))

login_email = driver.find_element(By.ID, 'passp-field-login')
login_email.send_keys(mail)

password_button = driver.find_element(By.XPATH, '//*[contains(@class, "passp-sign-in-button")]//button')
password_button.click()
WebDriverWait(driver,120).until(expected_conditions.presence_of_all_elements_located((By.ID, 'passp-field-passwd')))

passwords = driver.find_element(By.ID, 'passp-field-passwd')
passwords.send_keys(password)
WebDriverWait(driver,50)

button_2 = driver.find_element(By.XPATH, '//*[contains(@class, "passp-sign-in-button")]//button')
button_2.click()

WebDriverWait(driver,170).until(expected_conditions.presence_of_all_elements_located((By.XPATH, '//*[@data-testid = "left-column"]')))
WebDriverWait(driver,170).until(expected_conditions.presence_of_all_elements_located((By.XPATH, '//*[@data-testid = "messages-list_message-item_link"]')))
Mesages1 = driver.find_elements(By.XPATH, '//*[@data-testid = "messages-list_message-item"]')

messages1 = []
for message in Mesages1:
        message_link = message.find_element(By.XPATH, './/*[@data-testid = "messages-list_message-item_link"]').get_attribute('href')
        sender = message.find_element(By.XPATH, './/*[@data-testid = "message-common_sender-name"]//*[contains(@class, "Text")]').get_attribute('aria-label')
        message_list = message.find_element(By.XPATH, './/*[@data-testid = "messages-list_subject"]//*[contains(@class, "Text")]').text
        subject = message_list
        messages1.append({
            "message_link": message_link,
            "sender": sender,
            "subject": subject
        })

with open('messages1.json', 'w', encoding='utf-8') as f:
        json.dump(messages1, f, indent=2, ensure_ascii=False, default=str)

send_m_button = driver.find_element(By.XPATH, '//*[@href = "#sent"]')
send_m_button.click()

WebDriverWait(driver,120).until(expected_conditions.presence_of_all_elements_located((By.XPATH, '//*[@data-testid = "messages-list_message-item_link"]')))
Mesages2 = driver.find_elements(By.XPATH, '//*[@data-testid = "messages-list_message-item"]')

messages2 = []
for message in Mesages2:
        message_link = message.find_element(By.XPATH, './/*[@data-testid = "messages-list_message-item_link"]').get_attribute('href')
        sender = message.find_element(By.XPATH, './/*[@data-testid = "message-common_sender-name"]//*[contains(@class, "Text")]').get_attribute('aria-label')
        message_list = message.find_element(By.XPATH, './/*[@data-testid = "messages-list_subject"]//*[contains(@class, "Text")]').text
        subject = message_list
        messages2.append({
            "message_link": message_link,
            "sender": sender,
            "subject": subject
        })
with open('messages2.json', 'w', encoding='utf-8') as f:
        json.dump(messages2, f, indent=2, ensure_ascii=False, default=str)
        
driver.quit()
```
