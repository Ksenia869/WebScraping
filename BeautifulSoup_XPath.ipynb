# BeautifulSoup

```python
import pandas as pd
import requests
import lxml
from bs4 import BeautifulSoup
```

```python
url = "https://www.kinoafisha.info/releases/"
html = requests.post(url)
soup = BeautifulSoup(html.text, 'lxml')
print(html)
print(soup.title)
print(soup.h1)
```

```python
releases = soup.find_all('div', class_="movieList_title grid_cell12")
releases 
```

```python
list_relise = soup.find('div', class_="releases_list movieList movieList-grid grid")
list_movies = list_relise.find_all('div', class_="movieList_item movieItem movieItem-grid grid_cell3")
```

```python
films = []
for movies in list_movies:
    hrefs1 = movies.find('a', class_='movieItem_title')
    title = hrefs1.text

    ref1 = hrefs1['href'].replace('\n', '')
    r = requests.post(ref1)
    html_doc = r.text
    soup1 = BeautifulSoup(html_doc, 'lxml')
    description = soup1.find('span', class_='newFilmInfo_description')
    if description == None:
        description = ''
    else:
        description = description.text    

    title_secondary = soup1.find('span', class_='newFilmInfo_titleSecondary')
    if title_secondary == None:
        title_secondary = ''
    else:
        title_secondary = title_secondary.text    

    age = soup1.find('span', class_='newFilmInfo_age')
    if age == None:
        age = ''
    else:
        age = age.text 

    genre = soup1.find('nav', class_='newFilmInfo_genreMenu swipe outer-mobile inner-mobile')
    if genre == None:
        genre = ''
    else:
        genre = genre.get_text(', ',strip=True)

    producer = soup1.find('span', class_='badgeList_name')
    if producer == None:
        producer = ''
    else:
        producer = producer.text

    info_actors = soup1.find('div', class_='filmInfoActors_item filmInfoActors_item-actors')
    if info_actors == None:
        actors = ''
    else:
        actors = info_actors.find('div', class_='filmInfoActors_item_columns').get_text(', ',strip=True) 

    rating = soup1.find('div', class_='ratingBlockCard_local')
    if rating == None:
        rating = 0.0
    else:
        rating = rating.text

    info_films = soup1.find('div', class_='newFilmInfo_info swipe outer-mobile inner-mobile')
    if info_films == None:
        info_films = ''
    else:
        info_films = info_films.get_text(', ',strip=True)

    films.append([title, title_secondary, description, rating, age, genre, producer, actors, info_films])
    films_df = pd.DataFrame(films, columns= ['title', 'title_secondary', 'description', 'rating', 'age', 'genre', 'producer', 'actors', 'info_films'])
```

# XPath

```python
from lxml import html
```

```python
url = "https://www.kinoafisha.info/releases/"
r = requests.get(url)
```

```python
tree = html.fromstring(r.content)
```

```python
print(tree.xpath('.//title')[0].text_content().strip())
print(tree.xpath('.//h1')[0].text_content().strip())
```

```python
releases_xpath = tree.xpath('//div[@class="movieList_title grid_cell12"]')
for i in range(len(releases_xpath)):
    print(releases_xpath[i].text_content().strip())
```

```python
list_relise_xpath = tree.xpath('//div[@class="releases_list movieList movieList-grid grid"]')[0]
list_movies_xpath = list_relise_xpath.xpath('.//div[@class="movieList_item movieItem  movieItem-grid grid_cell3  "]')
```

```python
from lxml import html
import requests
import pandas as pd

films_xpath = []
for movies in list_movies_xpath:
    href_element = movies.xpath('.//a[@class="movieItem_title"]')[0]
    title = href_element.text_content().strip()
    ref1 = href_element.get('href').replace('\n', '')
    
    r = requests.post(ref1)
    tree = html.fromstring(r.content)
    
    description_elem = tree.xpath('//span[@class="newFilmInfo_description"]')
    description = description_elem[0].text_content().strip() if description_elem else ''
    
    title_secondary_elem = tree.xpath('//span[@class="newFilmInfo_titleSecondary"]')
    title_secondary = title_secondary_elem[0].text_content().strip() if title_secondary_elem else ''
    
    age_elem = tree.xpath('//span[@class="newFilmInfo_age"]')
    age = age_elem[0].text_content().strip() if age_elem else ''
    
    genre_elem = tree.xpath('//nav[@class="newFilmInfo_genreMenu swipe outer-mobile inner-mobile"]')
    if genre_elem:
        genre_texts = [text.strip() for text in genre_elem[0].xpath('.//text()') if text.strip()]
        genre = ', '.join(genre_texts)
    else:
        genre = ''
    
    producer_elem = tree.xpath('//span[@class="badgeList_name"]')
    producer = producer_elem[0].text_content().strip() if producer_elem else ''
    
    actors_elem = tree.xpath('//div[@class="filmInfoActors_item filmInfoActors_item-actors"]')
    if actors_elem:
        actors_columns = actors_elem[0].xpath('.//div[@class="filmInfoActors_item_columns"]')
        if actors_columns:
            actor_texts = [text.strip() for text in actors_columns[0].xpath('.//text()') if text.strip()]
            actors = ', '.join(actor_texts)
        else:
            actors = ''
    else:
        actors = ''
    
    rating_elem = tree.xpath('//div[@class="ratingBlockCard_local"]')
    rating = rating_elem[0].text_content().strip() if rating_elem else '0.0'
    
    info_films_elem = tree.xpath('//div[@class="newFilmInfo_info swipe outer-mobile inner-mobile"]')
    if info_films_elem:
        info_texts = [text.strip() for text in info_films_elem[0].xpath('.//text()') if text.strip()]
        info_films = ', '.join(info_texts)
    else:
        info_films = ''
    
    films_xpath.append([title, title_secondary, description, rating, age, genre, producer, actors, info_films])

films_xpath_df = pd.DataFrame(films_xpath, columns=['title', 'title_secondary', 'description', 'rating', 'age', 'genre', 'producer', 'actors', 'info_films'])
```
