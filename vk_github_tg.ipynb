```python
with open('key_gh.txt','r') as fil:
    token = fil.readline()

github = github3.login(token=token)  

all_repos =[]
for short_repository in github.repositories():
    repo_data = {
            'name': short_repository.name,
            'full_name': short_repository.full_name,
            'description': short_repository.description,
            'html_url': short_repository.html_url,
            'clone_url': short_repository.clone_url,
            'ssh_url': short_repository.ssh_url,
            'stars': short_repository.stargazers_count,
            'forks': short_repository.forks_count,
            'language': short_repository.language,
            'size': short_repository.size,
            'open_issues': short_repository.open_issues_count,
            'archived': short_repository.archived,
            'fork': short_repository.fork
        }
    all_repos.append(repo_data)
all_repos  

with open('repos_gh.json', 'w', encoding='utf-8') as f:
        json.dump(all_repos , f, indent=2, ensure_ascii=False, default=str)
```

```python
from telethon import TelegramClient
from telethon.tl.functions.channels import GetFullChannelRequest

with open('api_id.txt','r') as fil:
    api_id = fil.readline()

with open('api_hash.txt','r') as fil:
    api_hash = fil.readline() 

with open('phone.txt','r') as fil:
    phone = fil.readline() 

channel_username = '@obo_vsem_po_chyt_chytt'    
client = TelegramClient(phone, api_id, api_hash)

async with TelegramClient(phone, api_id, api_hash) as client:
    channel = await client.get_entity(channel_username)
    full_chat = await client(GetFullChannelRequest(channel))
    print(f"Название канала: {channel.title}")
    print(f"ID канала: {channel.id}")
    print(f"Username: @{channel.username}")
    print(f'Дата создания: {channel.date}')
    print(f"Количество подписчиков: {full_chat.full_chat.participants_count}")
    print(f"Описание: {full_chat.full_chat.about}")

    messages = await client.get_messages(channel_username , limit=20)

    print(f"Найдено сообщений: {len(messages)}\n")

    messages_data = []
        
    for message in messages:
            message_info = {
                'id': message.id,
                'date': message.date.isoformat() if message.date else None,
                'text': message.text,
                'media_type': str(message.media) if message.media else None
            }
            messages_data.append(message_info)
    
    export_data = {
            'chat': channel_username,
            'total_messages': len(messages_data),
            'messages': messages_data
        }

    filename = f'telegram_messages_{channel_username.replace("@", "")}.json'
    with open(filename, 'w', encoding='utf-8') as f:
            json.dump(export_data, f, ensure_ascii=False, indent=2) 

messages_data          
```
