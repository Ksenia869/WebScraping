```python
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi

client = MongoClient(uri, server_api=ServerApi('1'))
db = client['kinoafisha']

films = db.films.find()
for film in films:
    print(f"Название: {film.get('title')}")
    print(f"Дата выхода: {film.get('release_date')}")
    print(f"Жанры: {film.get('genres')}")
    print(f"Описание: {film.get('description')}")
    print("-" * 50)

count = db.films.count_documents({})
print(f"Всего фильмов в базе: {count}")
```
